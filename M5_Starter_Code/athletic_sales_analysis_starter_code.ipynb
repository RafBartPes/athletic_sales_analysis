```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
path_csv_2020 = ('Resources/athletic_sales_2020.csv')
path_csv_2021 = ('Resources/athletic_sales_2021.csv')

data_2020_df = pd.read_csv(path_csv_2020)
data_2021_df = pd.read_csv(path_csv_2021)
```

```python
# Display the 2020 sales DataFrame
data_2020_df.head()
```

```python
# Display the 2021 sales DataFrame
data_2021_df.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
data_2020_df.dtypes
```

```python
# Check the 2021 sales data types.
data_2021_df.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
data_2020_2021 = pd.concat([data_2020_df, data_2021_df], axis = "rows", join = "inner").reset_index(drop = True)
data_2020_2021.head()
```

```python
# Check if any values are null.
data_2020_2021.isnull().mean()
```

```python
# Check the data type of each column
data_2020_2021.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
data_2020_2021 = data_2020_2021.astype({"invoice_date" : "datetime64[ns]"})
```

```python
# Confirm that the "invoice_date" data type has been changed.
data_2020_2021.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".

data_2020_2021_gb= data_2020_2021.groupby(["region","state","city"])["units_sold"].sum().reset_index()
data_2020_2021_gb= data_2020_2021.rename(columns = {"units_sold":"Total_Products_Sold" })

# Show the top 5 results.
data_2020_2021_gb.head()                    
```

```python
# Show the number products sold for region, state, and city.

data_2020_2021_pt = pd.pivot_table(data_2020_2021, index = ["region","state","city"], values = "units_sold", aggfunc = ("sum"))


# Rename the "units_sold" column to "Total_Products_Sold"
data_2020_2021_pt= data_2020_2021.rename(columns = {"units_sold":"Total_Products_Sold" })

# Show the top 5 results.
data_2020_2021_pt.head(50)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
data_2020_2021_waf = data_2020_2021.loc[(data_2020_2021["product"] == "Women's Athletic Footwear")]
data_2020_2021_waf.head()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
data_2020_2021_waf_r = data_2020_2021_waf.groupby(['retailer','region', 'state', 'city'])[["units_sold"]].agg ("sum")

data_2020_2021_waf_r.rename(columns={"units_sold":"Womens_Footwear_Units_Sold"},inplace= True)

# Show the top 5 results.
data_2020_2021_waf_r.head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.

data_2020_2021_waf_r_pt = data_2020_2021_waf.pivot_table(values = "units_sold", index = ['retailer','region', 'state', 'city'], aggfunc = 'sum')
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
data_2020_2021_waf_r_pt.rename(columns={"units_sold":"Womens_Footwear_Units_Sold"},inplace = True)
# Show the top 5 results.
data_2020_2021_waf_r_pt.head()
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
data_2020_2021_waf_sales =  data_2020_2021_waf.pivot_table(values = "total_sales" , index = 'invoice_date', aggfunc = "sum")

# Optional: Rename the "total_sales" column to "Total Sales"
data_2020_2021_waf_sales.rename(columns = {"total_sales" : "Total Sales"}, inplace = True)

# Show the table.
data_2020_2021_waf_sales.head()
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
sales_resample_df = data_2020_2021_waf_sales.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
sales_resample_df.sort_values(by=["Total Sales"], ascending = True)
sales_resample_df.head()
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
sales_resample_weekly_df = data_2020_2021_waf_sales.resample('W').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
sales_resample_weekly_df.sort_values(by = ["Total Sales"], ascending = True)
sales_resample_weekly_df.head()
```
